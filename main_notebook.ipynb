## 1. Carbon Interface
#### API to calculate carbon (C02) emissions estimates for common C02 emitting activities

Information from the website:
- curl "https://www.carboninterface.com/api/v1/auth"
-  -H "Authorization: Bearer API_KEY"

```python
# https://docs.carboninterface.com/#/
```

```python
api_key = getpass.getpass()
```

```python
url = "https://www.carboninterface.com/api/v1/auth"
```

```python
# Testing a key
response = requests.get(url, headers = {'Authorization': f'Token {api_key}'})   # headers with authorization
```

```python
response.status_code
```

```python
url = "https://www.carboninterface.com/api/v1/estimates"
```

```python
payload = {
    "type": "vehicle",
    "distance_unit": "mi",
    "distance_value": 100,
    "vehicle_model_id": "7268a9b7-17e8-4c8d-acca-57059252afe9"
    }
```

```python
response = requests.post(url, json=payload, headers = {'Authorization': f'Token {api_key}',
                                       'Content-Type': 'application/json'})   # headers with authorization
```

```python
response.status_code
```

```python
results = response.json()
results
```

```python
data = pd.DataFrame(response.json())
data
```

```python
url = "https://www.carboninterface.com/api/v1/estimates"
payload = {
        "type": "electricity",
        "electricity_unit": "mwh",
        "electricity_value": 42,
        "country": "us",
        "state": "fl"
      }
```

```python
response.status_code
```

```python
results = response.json()
results
```
